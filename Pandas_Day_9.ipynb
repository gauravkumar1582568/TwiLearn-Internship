# Pandas Date Range

```python
# Year-Month-Date - Days Frequency 

pd.date_range(start='2022-02-02', end='2022-02-08')
```

```python
# Months Frequency 

pd.date_range(start='2022-02-01', end='2022-11-01', freq='M')
```

```python
# Weekdays Frequency 

pd.date_range(start='2022-02-01', end='2022-02-10', freq='B')
```

```python
data.columns = ['First Name','Current Age','College University','Location']
```

**Droping Columns**

```python
data.drop(['Current Age'], axis = 1)           # Temporary Droping Syntax
```

```python
data
```

```python
data.drop(['Current Age'], axis = 1, inplace = True)           # Permanent Droping Syntax
```

```python
data
```

```python
data
```

**Apply Method**

```python
# Creating Marks Column

data['Marks'] = [201, 210, 195, 183, 225, 203, 190]
```

```python
data
```

```python
# Apply Method

data['Marks'] = data['Marks'].apply(lambda x: x/5)
```

```python
data
```

```python
# Renaming Column

data.columns = ['First Name','College University','Location', 'Marks/Subject']
```

```python
data
```
